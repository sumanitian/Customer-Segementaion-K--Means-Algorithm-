# Clustering Process starts

### We are using with in cluster sum of square method to understand the number of optimum number of cluster we need

```python
X1 = df.loc[:,["Age","Spending Score (1-100)"]].values

from sklearn.cluster import KMeans
wcss = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k, init="k-means++")
    kmeans.fit(X1)
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(12,6))
plt.grid()
plt.plot(range(1,11),wcss, linewidth=2, color='green', marker = "8")
plt.xlabel("k-value")
plt.ylabel('wcss')
plt.show()
```

#### there is bend after 4 there is constant and we are considering number of cluster = 4

# So now we are going to form cluster based on age and spending score

```python
kmeans = KMeans(n_clusters=4)
label = kmeans.fit_predict(X1)
print(label)
```

##### Checking the centroids

```python
print(kmeans.cluster_centers_)
```

### Visualizing the clusters on graph
### drawing a scatter plot for that

```python
plt.scatter(X1[:,0], X1[:,1], c=kmeans.labels_, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], color='black')
plt.title('Clusters of Customers')
plt.xlabel('Age')
plt.ylabel('Spending Score(1-100)')
plt.show()
```

# Clusters based on Annual Income and Spending Score

```python
X2 = df.loc[:,["Annual Income (k$)","Spending Score (1-100)"]].values

from sklearn.cluster import KMeans
wcss = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k, init="k-means++")
    kmeans.fit(X2)
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(12,6))
plt.grid()
plt.plot(range(1,11),wcss, linewidth=2, color='red', marker = "8")
plt.xlabel("k-value")
plt.ylabel('wcss')
plt.show()
```

#### there is sharp bend around 5 and after that graph is smooth
#### number of cluster we are considering is 5

```python
kmeans = KMeans(n_clusters=5)
label = kmeans.fit_predict(X2)
print(label)
```

```python
print(kmeans.cluster_centers_)
```

```python
plt.scatter(X2[:,0], X2[:,1], c=kmeans.labels_, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], color='black')
plt.title('Annual Income')
plt.xlabel('Age')
plt.ylabel('Spending Score(1-100)')
plt.show()
```

# Considering all together age, annual income and spending score

```python
X3 = df.iloc[:,1:]
wcss = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k, init="k-means++")
    kmeans.fit(X3)
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(12,6))
plt.grid()
plt.plot(range(1,11),wcss, linewidth=2, color='orange', marker = "8")
plt.xlabel("k-value")
plt.ylabel('wcss')
plt.show()
```

#### after 5 it seems constant
#### considering 5 as a number of cluster

```python
kmeans = KMeans(n_clusters=5)
label = kmeans.fit_predict(X3)
print(label)
```

```python
print(kmeans.cluster_centers_)
```
